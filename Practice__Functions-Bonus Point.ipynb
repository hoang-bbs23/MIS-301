Define a function `discount_price(price, discount)`

###**Bonus Challenge**
Write a function called caculate_checkout that takes 3 inputs:
* **cart_total**: the total amount customer need to pay before any deductions or extras.
* **discount_code**: a coupon code ( a string). If discount_code is "Save10", apply 10% discount to the cart_total. If the discount_code is "Freeship", apply shipping cost $0.00. If there is no discount code applied, the discount_code by default is None.

* **shipping_cost**: the shipping cost by default is $5.00


**The function should return the final checkout amount after added any discount code and shipping cost.**

After define the function, check the following scenario:
1. Cart total $100, no discount code and default shipping

2. Cart total $60, used discount code Save10 and default shipping

3. Cart total $80, used discount code Freeshipping and shipping cost 6.00.

```python
# define function
def caculate_checkout(cart_total, discount_code = None, shipping_cost = 5.00):
  if discount_code == "Save10":
    cart_total = cart_total * 0.9                    # discount of 10%
  elif discount_code == "Freeship":
    shipping_cost = 0.0                              # no ship cost
  total_final = cart_total + shipping_cost
  return total_final

```

```python
# cart total 100, no discount code, default shipping
print(f"The final checkout = ${caculate_checkout(100)}")
```

```python
# cart total 60, save10 discount code, default shipping
print(f"The final checkout = ${caculate_checkout(60, "Save10")}")
```

```python
# cart total 80, freeship discount code, shipping cost $6.00
print(f"The final checkout = ${caculate_checkout(80, "Freeship", 6)}")
```
